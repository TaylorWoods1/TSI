# Spyder — IK & workspace demo

Requires the local package: `cd python && maturin develop --release`

````python
from spyder import Robot

r = Robot.rect(10, 6, 8)
lengths = r.ik(0.5, -0.2, 2.0)
print("IK lengths:", lengths)
pose = r.fk(lengths, 0, 0, 2)
print("FK recover:", pose)
````

````python
frac = r.workspace_fraction(-3, 3, -2, 2, 0.5, 5, 12, 10, 8)
print(f"Wrench-feasible fraction: {frac:.1%}")

traj = r.line_ik(0, 0, 2, 1.5, 0.5, 2.5, 12)
print(f"Trajectory samples: {len(traj)}, cables: {len(traj[0])}")
````
